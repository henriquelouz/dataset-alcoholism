```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('student-mat.csv', sep=',', skipinitialspace=True)
df2 = pd.read_csv('student-por.csv', sep=',', skipinitialspace=True)
df  = df1.merge(df2, on=["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"])
df
```

```python
df.columns
```

```python
df.drop('guardian_y', axis=1, inplace=True)
df.drop('higher_y', axis=1, inplace=True)
df.drop('romantic_y', axis=1, inplace=True)
df.drop('famrel_y', axis=1, inplace=True)
df.drop('freetime_y', axis=1, inplace=True)
df.drop('goout_y', axis=1, inplace=True)
df.drop('Dalc_y', axis=1, inplace=True)
df.drop('Walc_y', axis=1, inplace=True)
df.drop('health_y', axis=1, inplace=True)
df.rename(inplace=True, columns={'guardian_x':'guardian','higher_x':'higher','romantic_x':'romantic','famrel_x':'famrel','freetime_x':'freetime','goout_x':'goout','Dalc_x':'Dalc','Walc_x':'Walc','health_x':'health'})
df.columns
```

```python
from kmodes.kmodes import KModes
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cost = []
K = range(1,5)
for num_clusters in list(K):
    kmode = KModes(n_clusters=num_clusters, init="random", n_init=5)
    kmode.fit_predict(df)
    cost.append(kmode.cost_)

plt.plot(K, cost, 'bx-')
plt.xlabel('No. of clusters')
plt.ylabel('Cost')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmode = KModes(n_clusters=3, init = "random", n_init = 5)
clusters = kmode.fit_predict(df)
df.insert(0, "Cluster", clusters, True)
df
```
